## Filling Null Values

```python
df2 = df.fillna(value = 0)
df2.head()
```

```python
df2.isnull().sum().sum()
```

```python
# Fill null values with previous value
df3 = df.fillna(method = "pad")
df3
```

```python
df3.isnull().sum().sum()
```

```python
# Fill null values with next value
df4 = df.fillna(method = "bfill")
df4
```

```python
# Fill null values with previous value of Row
df5 = df.fillna(method = "pad", axis = 1)
df5
```

```python
# Fill null values with next value
df6 = df.fillna(method = "bfill",  axis  = 1)
df6
```

## Filling Diffrent Null  Values in diffrent culomns

```python
df7 = df.fillna({'society' :'abcd'})
df7
```

```python
df8 = df.fillna({'society':'abcd',
                'balcony': '234567' })
df8
```

## Fill the null values with mean

```python
df9 = df.fillna(value = df["balcony"].mean()) #we can do the same with max or min or ...
df9
```

## Dropna()
