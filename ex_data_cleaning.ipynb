```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.drop(columns = 'Not_Useful_Column')
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df['Last_Name'].str.strip('...')
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('...')
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('/')
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('_')
```

```python
df['Last_Name']
```

```python
df
```

```python
# df['Last_Name'] = df['Last_Name'].str.strip('_./') Or all in one thats possible too!
```

```python
# df['Phone_Number'].str.replace('^A-Za-z0-9','')
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^A-Za-z0-9]', '', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand=True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer']
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact']
```

```python
df
```

```python
df = df.replace('N/a', '')
```

```python
df = df.replace('NaN', '')
```

```python
df
```

```python
df = df.fillna('')
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
        
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', 'N')
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
        
df
```

```python
df = df.drop('Address', axis=1)
```

```python
df = df.reset_index()
df
```

```python
df = df.reset_index(drop=True)
df
```
